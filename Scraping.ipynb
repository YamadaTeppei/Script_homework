```python
import requests
import time
import pprint
from bs4 import BeautifulSoup

URL = 'https://news.yahoo.co.jp/topics/world?page=1'
response = requests.get(URL)
soup = BeautifulSoup(response.text, "html.parser")
topics = {}
    
j = 0
x = 1
while response:
    i = j + 1
    for tag, href in zip(soup('div', class_='newsFeed_item_title'), soup('a', class_='newsFeed_item_link')):
        topics[i] = {'title' : tag.text, 'href' : href.get('href')}
        i += 1
        time.sleep(0.5)
    j = i
    x += 1
    URL = 'https://news.yahoo.co.jp/topics/world?page=' +  str(x)
    response = requests.get(URL)

    
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(topics)
```
